```python
import pandas as pd
import matplotlib.pyplot as plt

# if you want nicer default styles:
plt.style.use("seaborn-v0_8-darkgrid")

DATA_PATH = "data_clean"
```

```python
def load_and_clean(fname):
    df = pd.read_csv(f"{DATA_PATH}/{fname}")
    # normalize to lower, snake_case, remove any weird chars
    df.columns = (
        df.columns
          .str.strip()
          .str.lower()
          .str.replace(" ", "_")
          .str.replace(r"[^\w_]", "", regex=True)
    )
    return df

strategic = load_and_clean("strategic_annual.csv")
hr        = load_and_clean("hr_annual.csv")
peers     = load_and_clean("peer_revenues.csv")
```

```python
# after cleaning, the column will be 'operating_margin_'
# (because parentheses and % got stripped); let's rename it:
strategic = strategic.rename(columns={
    "operating_margin_": "operating_margin_pct"
})
```

```python
print(strategic.columns.tolist())
# ['year', 'total_revenue_usd_million', 'operating_income_usd_million', 'operating_margin_pct']
```

```python
latest = int(strategic["year"].max())

# 4.1 Revenue growth %
strat_sorted = strategic.sort_values("year")
rev_growth = strat_sorted["total_revenue_usd_million"] \
                 .pct_change().iloc[-1] * 100

# 4.2 Operating margin % (just pull the latest)
op_margin = float(
    strategic.loc[strategic["year"] == latest, "operating_margin_pct"]
             .iloc[0]
)

# 4.3 Attrition rate % = voluntary_leavers / average_headcount
attr_rate = (
    hr.loc[hr["year"] == latest, "voluntary_leavers"].squeeze() /
    hr.loc[hr["year"] == latest, "average_headcount"].squeeze()
) * 100

print(f"For {latest}:")
print(f" • Rev growth:   {rev_growth:.1f}%")
print(f" • Op margin:    {op_margin:.1f}%")
print(f" • Attrition:    {attr_rate:.1f}%")
```

```python
fig, ax = plt.subplots(figsize=(8,4))

# plot Accenture line
ax.plot(
    strategic["year"],
    strategic["total_revenue_usd_million"],
    marker="o", lw=2, label="Accenture"
)

# plot each peer column
for col in peers.columns:
    if col == "year": continue
    ax.plot(peers["year"], peers[col], marker="o", lw=1.5, label=col.replace("_", " ").title())

ax.set_title("Annual Revenue: Accenture vs. Peers")
ax.set_xlabel("Year")
ax.set_ylabel("Revenue (USD million)")
ax.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from dash import Dash, dcc, html, dash_table, Input, Output
import dash_bootstrap_components as dbc
import plotly.express as px
```

```python
# — 1) Load Data —
strategic = pd.read_csv("data_clean/strategic_annual.csv")
hr        = pd.read_csv("data_clean/hr_annual.csv")
peers     = pd.read_csv("data_clean/peer_revenues.csv")
```

```python
# — 2) Pre-calc KPIs & options —
years = sorted(strategic["year"].unique())
latest = years[-1]
```

```python
# 2.1 Revenue Growth %
df_s = strategic.sort_values("year")
rev_growth = df_s["total_revenue_(usd_million)"].pct_change().iloc[-1] * 100
```

```python
# 2.2 Operating Margin %
op_margin = strategic.loc[strategic["year"]==latest, "operating_margin_(%)"].item()
```

```python
# 2.3 Attrition Rate %
vol = hr.loc[hr["year"]==latest, "voluntary_leavers"].item()
avg = hr.loc[hr["year"]==latest, "average_headcount"].item()
attr_rate = vol/avg * 100
```

```python
# 1) Standardize & inspect column names
hr.columns = (
    hr.columns
      .str.strip()                   # remove leading/trailing whitespace
      .str.lower()                   # lowercase everything
      .str.replace(" ", "_")        # spaces → underscores
      .str.replace(r"[^\w_]", "", regex=True)  # drop any punctuation
)
print("HR columns:", hr.columns.tolist())
```

```python
# using your cleaned hr DataFrame from above
hr["training_per_fte"] = hr["total_training_hours"] / hr["average_headcount"]
```

```python
# — 3) App & Layout —
app = JupyterDash(__name__, external_stylesheets=[dbc.themes.FLATLY])
app.title = "Accenture HR & Strategic Dashboard"
```

```python
# KPI cards
cards = dbc.Row([
    dbc.Col(dbc.Card([
        dbc.CardHeader("Revenue Growth %"),
        dbc.CardBody(html.H2(f"{rev_growth:.1f}%"))
    ], color="light")),
    dbc.Col(dbc.Card([
        dbc.CardHeader("Operating Margin %"),
        dbc.CardBody(html.H2(f"{op_margin:.1f}%"))
    ], color="light")),
    dbc.Col(dbc.Card([
        dbc.CardHeader("Attrition Rate %"),
        dbc.CardBody(html.H2(f"{attr_rate:.1f}%"))
    ], color="light")),
], className="mb-4")
```

```python
# Revenue comparison figure
fig_rev = px.line(
    pd.merge(strategic[["year","total_revenue_(usd_million)"]], peers, on="year"),
    x="year",
    y=[
      "total_revenue_(usd_million)",
      "ibm_(usd_million)",
      "capgemini_(eur_million)",
      "cognizant_(usd_million)"
    ],
    markers=True,
    labels={"value":"Revenue (USD m)","year":"Year","variable":"Company"},
    title="Annual Revenue: Accenture vs Peers"
)
fig_rev.update_layout(hovermode="x unified", legend_title_text="Company")
```

```python
# Attrition & Training figures
fig_attr = px.line(
    hr, x="year", y=hr["voluntary_leavers"]/hr["average_headcount"],
    markers=True,
    labels={"y":"Attrition Rate","year":"Year"},
    title="Annual Attrition Rate"
)
fig_attr.update_yaxes(tickformat=".1%")
fig_train = px.line(
    hr, x="year", y="training_per_fte",
    markers=True,
    labels={"training_per_fte":"Training Hours per FTE","year":"Year"},
    title="Annual Training Hours per FTE"
)

layout = dbc.Container([
    html.H1("🔹 Accenture Strategic & HR Dashboard", className="my-4"),
    cards,
    dbc.Row([
        dbc.Col(dcc.Graph(figure=fig_rev), width=12)
    ]),
    html.Hr(),
    dbc.Row([
        dbc.Col(dcc.Graph(figure=fig_attr), width=6),
        dbc.Col(dcc.Graph(figure=fig_train), width=6),
    ]),
    html.Hr(),
    html.H4("Drill-down HR Details"),
    dbc.Row([
        dbc.Col(
            dcc.Dropdown(
                id="year-dropdown",
                options=[{"label": y, "value": y} for y in years],
                value=latest,
                clearable=False
            ), width=4
        )
    ], className="mb-3"),
    dash_table.DataTable(
        id="hr-table",
        columns=[{"name": c, "id": c} for c in hr.columns],
        style_table={"overflowX":"auto"},
        style_cell={"padding":"5px","textAlign":"left"},
    )
], fluid=True)

app.layout = layout
```

```python
import plotly.express as px

# 1) Operating Income Over Time
fig_income = px.bar(
    strategic,
    x="year",
    y="operating_income_(usd_million)",       # ← here
    title="Operating Income (USD m) Over Time",
    labels={
      "year": "Year",
      "operating_income_(usd_million)": "Op. Income (USD m)"
    }
)


# 2) Headcount Trend
fig_headcount = px.line(
    hr,
    x="year",
    y="average_headcount",
    markers=True,
    title="Average Headcount Over Time",
    labels={"year":"Year", "average_headcount":"Avg. Headcount"}
)

# 3) Engagement Score Trend
fig_engage = px.line(
    hr,
    x="year",
    y="engagement_score",
    markers=True,
    title="Employee Engagement Score Over Time",
    labels={"year":"Year", "engagement_score":"Engagement Score"}
)
```

```python
from dash import Input, Output, dash_table
# assume you've already defined:
#   app = dash.Dash(...)
#   hr  = <your pandas DataFrame>
#   years = sorted(hr.year.unique())

# 1) add the dropdown and empty table to your layout:
app.layout = dbc.Container([
    # ... your KPIs and graphs above ...

    html.H4("Drill-down HR Details"),
    dbc.Row([
        dbc.Col(dcc.Dropdown(
            id="year-dropdown",
            options=[{"label":y, "value":y} for y in years],
            value=latest,
            clearable=False
        ), width=4),
    ], className="mb-3"),
    dash_table.DataTable(
        id="hr-table",
        columns=[{"name":c, "id":c} for c in hr.columns],
        page_size=10,
        style_table={"overflowX":"auto"},
        style_cell={"padding":"5px","textAlign":"left"},
    )
], fluid=True)
@app.callback(
  Output("rev-vs-peers", "figure"),
  Input("peer-selector", "value"),
)
def update_rev_peers(peers_selected):
    # always include Accenture’s total_revenue
    df = pd.merge(
      strategic[["year","total_revenue_usd_million"]],
      peers[["year"] + peers_selected],
      on="year", how="left"
    )
    fig = px.line(
      df, x="year",
      y=["total_revenue_usd_million"] + peers_selected,
      markers=True,
      title="Annual Revenue: Accenture vs Peers",
      labels={"value":"Revenue (USD m)","variable":"Company"}
    )
    return fig
fig_rev.update_traces(selector=dict(name="total_revenue_usd_million"), 
                      line=dict(color="blue"))
fig_rev.update_traces(selector=dict(name="ibm_usd_million"), line=dict(color="orange"))
# …and so on, or rely on Plotly’s default color cycle—but remove the black default
fig_rev.update_layout(legend_title_text="Company")


# 2) register a callback to filter hr by the selected year:
@app.callback(
    Output("hr-table", "data"),
    Input("year-dropdown", "value")
)
def update_hr_table(selected_year):
    df = hr[hr["year"] == selected_year]
    return df.to_dict("records")
@app.callback(
  Output("forecast-chart","figure"),
  Input("horizon-slider","value"),
  Input("peer-selector","value")
)
def update_forecast(horizon, peers_selected):
    # merge base data
    df = pd.merge(
      strategic[["year","total_revenue_usd_million"]],
      peers[["year"] + peers_selected],
      on="year", how="left",
    )
    # simple linear projection: carry last growth rate forward
    last_growth = df["total_revenue_usd_million"].pct_change().iloc[-1]
    future_years = list(range(df.year.max()+1, df.year.max()+horizon+1))
    last_val = df.total_revenue_usd_million.iloc[-1]
    proj = [ last_val * (1+last_growth)**i for i in range(1,horizon+1) ]
    df = df.append(pd.DataFrame({
      "year": future_years,
      "total_revenue_usd_million": proj
    }), ignore_index=True)
    fig = px.line(
      df, x="year", y="total_revenue_usd_million",
      markers=True,
      title=f"{horizon}-Year Forecast: Accenture vs Peers"
    )
    return fig
fig_rev.update_yaxes(tickformat=",")             # shows “10,000” instead of “10000”
fig_attr.update_yaxes(tickformat=".1%")          # shows “11.2%”
fig_train.update_yaxes(tickformat=".0f")         # hours per FTE

```

```python
import dash_bootstrap_components as dbc
from dash import html

# --- 4) Build KPI cards ----------------------------------

# 4.1: Assemble a list of (title, value, color) triples
kpi_data = [
    ("Revenue Growth %", f"{rev_growth:.1f}%", "primary"),
    ("Op. Margin %",      f"{op_margin:.1f}%",    "success"),
    ("Attrition Rate %",   f"{attr_rate:.1f}%",    "danger"),
]

# 4.2: For each KPI, make a little card
kpi_cards = dbc.Row(
    [
        dbc.Col(
            dbc.Card(
                dbc.CardBody(
                    [
                        html.H6(title, className="card-title"),
                        html.H3(value, className="card-text"),
                    ]
                ),
                color=color,
                inverse=True,
                className="mb-4",
            ),
            width=4,
        )
        for title, value, color in kpi_data
    ],
    className="mb-4",
)

# --- 5) Final layout ------------------------------------

app.layout = dbc.Container(
    [
        html.H1("⚙️ Accenture Strategic & HR Dashboard", className="text-center my-4"),

        # KPI cards row (now defined!)
        kpi_cards,

        # Row 1: Revenue + Income
        dbc.Row(
            [
                dbc.Col(dcc.Graph(figure=fig_rev),    width=6),
                dbc.Col(dcc.Graph(figure=fig_income), width=6),
            ],
            className="mb-4",
        ),

        # Row 2: Attrition + Training + Engagement
        dbc.Row(
            [
                dbc.Col(dcc.Graph(figure=fig_attr),   width=4),
                dbc.Col(dcc.Graph(figure=fig_train),  width=4),
                dbc.Col(dcc.Graph(figure=fig_engage), width=4),
            ],
            className="mb-4",
        ),

        # Row 3: Headcount Trend
        dbc.Row(
            [
                dbc.Col(dcc.Graph(figure=fig_headcount), width=12),
            ],
            className="mb-4",
        ),

        html.Hr(),

        # Row 4: Drill-down HR table
        html.H4("Drill-down HR Details"),
        dbc.Row(
            [
                dbc.Col(
                    dcc.Dropdown(
                        id="year-dropdown",
                        options=[{"label": y, "value": y} for y in years],
                        value=latest,
                        clearable=False,
                    ),
                    width=3,
                ),
            ],
            className="mb-3",
        ),
        dash_table.DataTable(
            id="hr-table",
            columns=[{"name": c, "id": c} for c in hr.columns],
            page_size=10,
            style_table={"overflowX": "auto"},
            style_cell={"padding": "5px", "textAlign": "left"},
        ),
    ],
    fluid=True,
)
    # ------------------------------------------------
   
      
  

```
